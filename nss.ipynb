```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load data
df = pd.read_csv("nss2025.csv")

# Question metadata
questions = {
    1: {"question": "Q01: How good are teaching staff at explaining things?", "category": "Teaching"},
    2: {"question": "Q02: How often do teaching staff make the subject engaging?", "category": "Teaching"},
    3: {"question": "Q03: How often is the course intellectually stimulating?", "category": "Course"},
    4: {"question": "Q04: How often does your course challenge you to achieve your best work?", "category": "Course"},
    5: {"question": "Q05: To what extent have you had the chance to explore ideas and concepts in depth?", "category": "Course"},
    6: {"question": "Q06: How well does your course introduce subjects and skills in a way that builds on what you have already learned?", "category": "Course"},
    7: {"question": "Q07: To what extent have you had the chance to bring together information and ideas from different topics?", "category": "Course"},
    8: {"question": "Q08: To what extent does your course have the right balance of directed and independent study?", "category": "Course"},
    9: {"question": "Q09: How well has your course developed your knowledge and skills that you think you will need for your future?", "category": "Course"},
    10: {"question": "Q10: How clear were the marking criteria used to assess your work?", "category": "Exams and marking"},
    11: {"question": "Q11: How fair has the marking and assessment been on your course?", "category": "Exams and marking"},
    12: {"question": "Q12: How well have assessments allowed you to demonstrate what you have learned?", "category": "Exams and marking"},
    13: {"question": "Q13: How often have you received assessment feedback on time?", "category": "Exams and marking"},
    14: {"question": "Q14: How often does feedback help you to improve your work?", "category": "Exams and marking"},
    15: {"question": "Q15: How easy was it to contact teaching staff when you needed to?", "category": "Teaching"},
    16: {"question": "Q16: How well have teaching staff supported your learning?", "category": "Teaching"},
    17: {"question": "Q17: How well organised is your course?", "category": "Course"},
    18: {"question": "Q18: How well were any changes to teaching on your course communicated?", "category": "Communications"},
    19: {"question": "Q19: How well have the IT resources and facilities supported your learning?", "category": "Facilities"},
    20: {"question": "Q20: How well have the library resources (e.g., books, online services and learning spaces) supported your learning?", "category": "Facilities"},
    21: {"question": "Q21: How easy is it to access subject specific resources (e.g., equipment, facilities, software) when you need them?", "category": "Facilities"},
    22: {"question": "Q22: To what extent do you get the right opportunities to give feedback on your course?", "category": "Feedback"},
    23: {"question": "Q23: To what extent are students' opinions about the course valued by staff?", "category": "Feedback"},
    24: {"question": "Q24: How clear is it that students' feedback on the course is acted on?", "category": "Feedback"},
    25: {"question": "Q25: How well does the students' union (association or guild) represent students' academic interests?", "category": "Feedback"},
    26: {"question": "Q26: How well communicated was information about your university/college's mental wellbeing support services?", "category": "Communications"},
    27: {"question": "Q27: During your studies, how free did you feel to express your ideas, opinions, and beliefs?", "category": "Academic freedom"},
}

# Build lookup: question -> category
question_category_map = {
    q["question"]: q["category"]
    for q in questions.values()
}

# Attach category to dataframe
df["category"] = df["question"].map(question_category_map)

df_filtered = df[df["cah3_subject"] == "medicine (non-specific)"]

# Group by category and calculate mean agreement
category_summary = (
    df_filtered.groupby("category", as_index=False)
      .agg(mean_agree_pct=("agree_pct", "mean"))
      .sort_values("mean_agree_pct", ascending=True)
)

# Prepare plot data
labels = category_summary["category"].tolist()
values = category_summary["mean_agree_pct"].tolist()

# Plot
fig, ax = plt.subplots(figsize=(12, 8))
ax.barh(labels, values)
ax.set_xlabel("Percentage")
ax.set_title("NSS Results by Category")
ax.grid(axis="x", linestyle="--", alpha=0.7)
ax.invert_yaxis()

plt.tight_layout()
plt.show()
```
